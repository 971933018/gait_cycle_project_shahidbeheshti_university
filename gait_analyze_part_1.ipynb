```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
data  = pd.read_excel('/content/drive/MyDrive/data_project/5day_night_23_45')
```

```python
plt.plot(data['Time (s)'] , data['Linear Acceleration x (m/s^2)'] , label ='accelearation x ')
#plt.plot(data['Time (s)'] , data['Linear Acceleration y (m/s^2)'] , label ='accelearation y ')
#plt.plot(data['Time (s)'] , data['Linear Acceleration z (m/s^2)'] , label ='accelearation z ')
plt.plot(data['Time (s)'] , data['Absolute acceleration (m/s^2)'] , label ='accelearation absolute ')


plt.legend()
plt.grid()
plt.show()
```

```python
data['a_x_rolling'] = data['Linear Acceleration x (m/s^2)'].rolling(15).mean()
```

```python
plt.plot(data['Time (s)'] , data['a_x_rolling'] , label ='rolling accelearation x ')
plt.plot(data['Time (s)'] , data['Linear Acceleration x (m/s^2)'] , label ='accelearation x ')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(data['Time (s)'] , data['a_x_rolling'] , label ='rolling accelearation x ')
plt.show()
```

```python
plt.plot(data['Time (s)'] , data['Linear Acceleration x (m/s^2)'] , label ='accelearation x ')
plt.show()
```

```python
from scipy.signal import find_peaks 
```

```python
list_peaks_x = find_peaks(data['a_x_rolling'])
```

```python
plt.scatter(data['Time (s)'][list_peaks_x[0]] , data['a_x_rolling'][list_peaks_x[0]] , label = 'peaks'  , c = 'r')
plt.plot(data['Time (s)'] , data['a_x_rolling'] , label= ' linear x acceleration ')
plt.title( ' data Acceleration rolling X with peaks ')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(data['Time (s)'] , data['Linear Acceleration y (m/s^2)'] , label = ' linear a - y ')
plt.legend()
plt.grid()
plt.show()
```

```python
data['a_y_rolling'] = data['Linear Acceleration y (m/s^2)'].rolling(15).mean()
```

```python
plt.plot(data['Time (s)'] , data['a_y_rolling'] , label = 'a - y rolling ')
plt.title(' rolling a - y ')
plt.xlabel(' time ')
plt.ylabel(' accelearation rooling y ')
plt.legend()
plt.grid()
plt.show()
```

```python
list_peaks_y = find_peaks(data['a_y_rolling'])
```

```python
plt.plot(data['Time (s)'] , data['a_y_rolling'] , label = ' a -y rolling ')
plt.scatter(data['Time (s)'][list_peaks_y[0]] , data['a_y_rolling'][list_peaks_y[0]] , label = ' peaks a _ y rolling ' , c = 'r')
plt.legend()
plt.title(' data acceleration y rolling with peaks ')
plt.grid()
plt.show()
```
