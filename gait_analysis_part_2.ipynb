```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt  
```

```python
data = pd.read_excel('/content/drive/MyDrive/data_project/5day_night_23_45')
```

```python
data.head()
```

```python
data['rolling_x'] = data['Linear Acceleration x (m/s^2)'].rolling(15).mean()
data['rolling_y'] = data['Linear Acceleration y (m/s^2)'].rolling(15).mean()
data['rolling_z'] = data['Linear Acceleration z (m/s^2)'].rolling(15).mean() 
```

```python
data['absolute_rolling'] = data['Absolute acceleration (m/s^2)'].rolling(15).mean()
```

```python
fig = plt.figure(figsize = (10, 10 ))
plt.plot(data['Time (s)'] , data['rolling_x'] , label = 'rolling x ')
plt.plot(data['Time (s)'] , data['rolling_y'] , label = 'rolling y ')
plt.plot(data['Time (s)'] , data['rolling_z'] , label = 'rolling z ')
plt.legend()
plt.grid()
plt.xlabel(' Time of data collecting ')
plt.ylabel(' rolling acceleration in 3 D axis ')
plt.show()
```

```python
from scipy.signal import find_peaks  
```

```python
list_peaks_x = find_peaks(data['rolling_x'])
list_peaks_y = find_peaks(data['rolling_y'])
list_peaks_z = find_peaks(data['rolling_z'])
list_peaks_abs = find_peaks(data['absolute_rolling']) 
```

```python
fig , ax = plt.subplots( 4  , figsize = (10, 10 ))
ax[0].plot(data['Time (s)'] , data['rolling_x'] , label = 'rolling x ')
ax[0].scatter(data['Time (s)'][list_peaks_x[0]] , data['rolling_x'][list_peaks_x[0]] , c = 'r')
ax[0].plot(data['Time (s)'][list_peaks_x[0]] , data['rolling_x'][list_peaks_x[0]] , c = 'green')
ax[0].legend()
ax[0].grid()





ax[1].plot(data['Time (s)'] , data['rolling_y'] , label = 'rolling y ')
ax[1].scatter(data['Time (s)'][list_peaks_y[0]] , data['rolling_y'][list_peaks_y[0]] , c = 'r')
ax[1].plot(data['Time (s)'][list_peaks_y[0]] , data['rolling_y'][list_peaks_y[0]] , c = 'green')
ax[1].legend()
ax[1].grid()


ax[2].plot(data['Time (s)'] , data['rolling_z'] , label = 'rolling z ')
ax[2].scatter(data['Time (s)'][list_peaks_z[0]] , data['rolling_z'][list_peaks_z[0]] , c = 'r')
ax[2].plot(data['Time (s)'][list_peaks_z[0]] , data['rolling_z'][list_peaks_z[0]] , c = 'gold')
ax[2].legend()
ax[2].grid()


ax[3].plot(data['Time (s)'] , data['absolute_rolling'] , label = 'absolute_rolling')
ax[3].scatter(data['Time (s)'][list_peaks_abs[0]] , data['absolute_rolling'][list_peaks_abs[0]] , c = 'r' )
ax[3].plot(data['Time (s)'][list_peaks_abs[0]] , data['absolute_rolling'][list_peaks_abs[0]] , c = 'gold' )
ax[3].legend()
ax[3].grid()



plt.show()

```
